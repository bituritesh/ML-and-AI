### Starting with Date_of_Journey format is object -> date time format. We need to create 3 diff features for dd, mm, yyyy


```python
# df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
# df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
# df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
# or
# cols=['date','month','year']for i,col in enumerate(cols): df[col]=df['Date_of_Journey'].apply(lambda x : int(x.split('/')[i]))
# or
df["Date"] = df['Date_of_Journey'].apply(lambda x:x.split("/")[0])
df["Month"] = df['Date_of_Journey'].apply(lambda x:x.split("/")[1])
df["Year"] = df['Date_of_Journey'].apply(lambda x:x.split("/")[2])
print(df['Date'],df['Month'],df['Year'])
```

```python
df.head()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df.info()
```

```python
df.drop('Date_of_Journey', inplace=True,axis=1)
```

```python
df.head(1)
```

```python
df['Arrival_Time'] =df['Arrival_Time'].apply(lambda x:x.split()[0])
df.head()
```

```python
df['Arrival_Hours'] =df['Arrival_Time'].apply(lambda x:x.split(':')[0]).astype(int)
df['Arrival_Minutes'] =df['Arrival_Time'].apply(lambda x:x.split(':')[1]).astype(int)
df.drop('Arrival_Time', inplace=True,axis=1)
df.head()
```
